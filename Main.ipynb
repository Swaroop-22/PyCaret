```python
#Import libraries
import pandas as pd
from pycaret.classification import *
```

# Load Data

```python
data = pd.read_csv('heart.csv')
```

# Train And Evaluate Model

```python
cat_feature = ['sex','cp', 'fbs','exang', 'thal']
```

setup() function is from PyCaret, it initiates the ML experiments and setups the training pipeline.
There are number of parameters that can be set for the experiment within the function.

```python
experiment = setup(data, target='condition', categorical_features=cat_feature)
```

In PyCaret, the compare_models() after running this command it comapre all the machine learning model and highlights the best suited model for your datast.

In this Case the Navie Bayes ML algorithms is best suited with Accuracy = 0.8445

```python
best_model = compare_models()
```

# Testing The Model

```python
predict_model(best_model, data.tail())
```

# Saving The Model

```python
save_model(best_model, model_name='Navie-Bayes')
```
